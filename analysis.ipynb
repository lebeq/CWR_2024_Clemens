```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
def line(arg, m, c):
    #straight line y = m*arg + c
    return m*arg+c
data = np.loadtxt('data_midpoint.csv', delimiter=',')
xdata = data[:,0]
ydata = data[:,1]
plt.loglog(data[:,0],data[:,1])
plt.show()
```

```python
fit = curve_fit(line,xdata,ydata)
popt, pcov = fit
popt
```
